# Pipeline Process

In order to process the images, I used the following approach:
1. Grayscale the image
2. Gaussian blurring 
3. Canny edge detection and selecting region of interest
4. Hough lines
5. Plotting the lines on the images
6. Identifying slopes and intercepts for each line to determine lanes

In order to calculate the slope and intercepts, I modified the "draw lines" function and performed a linear regression to interpolate the line.

I had moderate success with images and the white lines, but the yellow could be improved.

# Potential Improvements

This process could be improved in several ways:
1. Limiting the lines I considered to the most likely fit, either by observing the standard deviation or arbitrarily limiting the results. This would limit the "jumpiness" of the lines and result in a potentially smoother result. 
2. Pursuing nonlinear regression in the event the road has any imperfections or nonlinear representations of the lane lines
